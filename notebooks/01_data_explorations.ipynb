Bảng thống kê mô tả cung cấp các thước đo trung tâm và độ phân tán (mean, median, standard deviation) cùng với hệ số lệch (skewness) cho các biến số.

Nhìn chung, phần lớn các biến có phân phối không đối xứng, với nhiều giá trị tập trung ở mức thấp và một số ít giá trị lớn kéo đuôi phân phối về phía phải.

- **Các biến liên quan đến số lượng khách** (`adults`, `children`, `babies`) cho thấy đa số các lượt đặt phòng chỉ có 1-2 người lớn và rất ít trẻ em hoặc em bé, mặc dù vẫn tồn tại một số giá trị bất thường.

- **Các biến lịch sử đặt phòng** (`previous_cancellations`, `previous_bookings_not_canceled`, `booking_changes`) có median bằng 0 và skew rất cao, cho thấy phần lớn khách hàng không có lịch sử hoặc thay đổi đặt phòng trước đó.

- **Biến giá phòng (`adr`)** có phân phối lệch phải mạnh với một số giá trị rất lớn, đồng thời xuất hiện giá trị âm, cho thấy cần được kiểm tra và xử lý trong bước preprocessing.

Nhìn chung, đối với hầu hết các biến, **median nhỏ hơn mean**, cho thấy rằng phân phối bị kéo lệch bởi các giá trị lớn bất thường. Vì vậy, trong các phân tích tiếp theo, **median và IQR** sẽ là những thước đo trung tâm và độ phân tán phù hợp hơn so với mean và standard deviation.

##### Distribution Visualizations
###### Histogram  
Histogram được sử dụng để quan sát phân phối tổng quát của toàn bộ các biến số. Thông qua các biểu đồ này, có thể nhanh chóng nhận diện mức độ tập trung giá trị, hình dạng phân phối và sự xuất hiện của các đuôi phân phối dài.

```python
n_cols = 4
n_rows = math.ceil(len(num_cols) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 16))
axes = axes.ravel()

for i, col in enumerate(num_cols):
    axes[i].hist(df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
    axes[i].set_title(col, fontweight='bold', fontsize=10)
    axes[i].tick_params(axis='both', labelsize=8)

plt.suptitle("Distribution of Numerical Variables", fontsize=18, fontweight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

##### Outlier Identification
Phân tích outliers chỉ được áp dụng cho một số **biến liên tục có phạm vi giá trị rộng**, nơi các giá trị cực đoan có thể ảnh hưởng đáng kể đến phân tích. Các biến nhị phân, biến thời gian dạng lịch và biến đếm có miền giá trị nhỏ không được đưa vào do việc xác định outliers trong các trường hợp này không mang nhiều ý nghĩa.

Để xác định các giá trị ngoại lệ, sử dụng **box plot** để quan sát trực quan và **phương pháp IQR** để định lượng outliers. Hai phương pháp này phù hợp với dữ liệu có phân phối không đối xứng.

**Box Plot**

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

cols = [
    "lead_time",
    "stays_in_week_nights",
    "days_in_waiting_list",
    "adr"
]

for ax, col in zip(axes.ravel(), cols):
    sns.boxplot(y=df[col], ax=ax, showfliers=False, width=0.4)
    ax.set_title(col, fontweight="bold")
    ax.set_ylabel("")

plt.suptitle("Box Plots of Selected Numerical Variables (Outliers Hidden)", fontsize=12, fontweight="bold")
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

Các box plot cho thấy sự khác biệt rõ rệt về phân bố trung tâm (median) và mức độ phân tán (IQR) giữa các biến được phân tích.

- Ngược lại, `stays_in_week_nights` tập trung hơn quanh median, cho thấy thời gian lưu trú trong tuần khá ổn định.
- Đối với `days_in_waiting_list`, phần lớn giá trị bằng 0 khiến box plot bị thu hẹp thành một đường gần như phẳng. Điều này phản ánh đặc điểm zero-inflated của dữ liệu hơn là lỗi trong quá trình trực quan hóa.

Nhìn chung, box plot được sử dụng để mô tả phân bố trung tâm của dữ liệu, trong khi việc xác định outliers được thực hiện định lượng bằng phương pháp IQR ở bước tiếp theo.

**IQR-based Outlier Detection**

Phương pháp Interquartile Range (IQR) được sử dụng để xác định các giá trị ngoại lệ dựa trên ngưỡng thống kê, phù hợp với dữ liệu có phân phối không đối xứng.

Với mỗi biến:

- $IQR = Q3 - Q1$

- Ngưỡng dưới: $Q1 - 1.5 \times IQR$

- Ngưỡng trên: $Q3 + 1.5 \times IQR$

Các giá trị nằm ngoài khoảng này được xem là outliers.

```python
outlier_df = summarize_outliers_iqr(df, cols)
outlier_df
```
